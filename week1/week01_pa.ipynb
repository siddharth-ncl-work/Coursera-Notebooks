```python
def expand(X):
    X_expanded=np.zeros((X.shape[0],6))
    X_expanded[:,0:2]=X
    X_expanded[:,2:4]=X**2
    X_expanded[:,4]=X[:,0]*X[:,1]
    X_expanded[:,5]=1
    return X_expanded
    
```

```python
X_expanded=expand(X)
X_expanded[:,-1]
```

```python
dummy_X=np.array([[0,0],
                  [1,0],
                  [2.61,-1.28],
                  [-0.59,2.1]])

dummy_expanded_ans = np.array([[ 0.    ,  0.    ,  0.    ,  0.    ,  0.    ,  1.    ],
                               [ 1.    ,  0.    ,  1.    ,  0.    ,  0.    ,  1.    ],
                               [ 2.61  , -1.28  ,  6.8121,  1.6384, -3.3408,  1.    ],
                               [-0.59  ,  2.1   ,  0.3481,  4.41  , -1.239 ,  1.    ]])

dummy_expanded=expand(dummy_X)
print(dummy_expanded.shape)
print(np.allclose(dummy_expanded,dummy_expanded_ans,1e-3))
```

```python
def probability(X,w):
    return 1/(1+np.exp(-1*X@w))
```

```python
dummy_weights=np.linspace(-1,1,6)
probability(X_expanded[:1,:],dummy_weights)
```

```python
probability(X_expanded[:5,:],dummy_weights)
```

```python
def compute_loss(X,y,w):
    '''
    cross-entropy
    '''
    y_pred=probability(X,w)
    return -np.sum(y*np.log(y_pred)+(1-y)*np.log(1-y_pred))/len(y)
```

```python
compute_loss(X_expanded,y,dummy_weights)
```

```python
def compute_grad(X,y,w):
    y_pred=probability(X,w)
    return ((y_pred-y)@X)/len(y)
```

```python
np.linalg.norm(compute_grad(X_expanded,y,dummy_weights))
```

```python
from IPython import display

h=0.01
x_min, x_max = X[:,0].min()-1, X[:,0].max()+1
y_min, y_max = X[:,1].min()-1, X[:,1].max()+1

xx,yy=np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))

def visualize(X,y,w,history):
    Z=probability(expand(np.c_[xx.ravel(),yy.ravel()]),w)
    Z=Z.reshape(xx.shape)
    plt.subplot(1,2,1)
    plt.contourf(xx,yy,Z,alpha=0.8)
    plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.Paired)
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
    
    plt.subplot(1,2,2)
    plt.plot(history)
    plt.grid()
    ymin,ymax=plt.ylim()
    plt.ylim(0,ymax)
    display.clear_output(wait=True)
    plt.show()
```

```python
visualize(X,y,dummy_weights,[0.5,0.5,0.25])
```

```python
np.random.seed(42)
w=np.array([0,0,0,0,0,1])
eta=0.1

n_iter=100
batch_size=4
loss=np.zeros(n_iter)
plt.figure(figsize=(12,5))

for i in range(n_iter):
    ind=np.random.choice(X_expanded.shape[0],batch_size)
    loss[i]=compute_loss(X_expanded,y,w)
    if i%10==0:
        visualize(X_expanded[ind,:],y[ind],w,loss)
    grad=compute_grad(X_expanded[ind,:],y[ind],w)
    w=w-eta*grad
visualize(X,y,w,loss)
plt.clf()
```
